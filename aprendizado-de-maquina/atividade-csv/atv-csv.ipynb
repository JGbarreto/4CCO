```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
campeonato = pd.read_csv("campeonato-brasileiro-estatisticas-full - campeonato-brasileiro-estatisticas-full.csv", delimiter=',')
```

```python
faltas_por_time = campeonato.groupby('clube')['faltas'].sum().reset_index()
norm = plt.Normalize(min(campeonato['faltas']), max(campeonato['faltas']))

colors = plt.cm.coolwarm(norm(campeonato['faltas']))
posicoes = np.arange(len(campeonato)) * 1.5 

plt.bar(faltas_por_time['clube'], faltas_por_time['faltas'], color=colors)

plt.title('Total de faltas por time')
plt.xlabel('Clube')
plt.ylabel('Qtd. Faltas')
plt.xticks(rotation=90, fontsize=8) 
```
