```python

from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1)
X, y = mnist["data"], mnist["target"]
print(X.shape)
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.svm import SVC

svcL = SVC(kernel='linear')
svcP = SVC(kernel='poly')
svcR = SVC(kernel='rbf')
```

```python
svcL.fit(X_train_scaled, y_train)
svcP.fit(X_train_scaled, y_train)
svcR.fit(X_train_scaled, y_train)
```

```python
y_predL = svcL.predict(X_test_scaled)
y_predP = svcP.predict(X_test_scaled)
y_predR = svcR.predict(X_test_scaled)   
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print("Linear Kernel")
print(confusion_matrix(y_test, y_predL))
print(classification_report(y_test, y_predL))
```

```python
print("Polynomial Kernel")
print(confusion_matrix(y_test, y_predP))
print(classification_report(y_test, y_predP))

```

```python
print("RBF Kernel")
print(confusion_matrix(y_test, y_predR))
print(classification_report(y_test, y_predR))   
```
